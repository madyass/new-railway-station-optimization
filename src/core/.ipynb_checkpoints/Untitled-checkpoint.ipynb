```python
import folium
from shapely.geometry import Polygon, Point
import numpy as np

# İstanbul poligonu koordinatları
istanbul_coords = [
    [41.00315374563969, 28.498494947567604],
    [40.958055520667095, 28.824651557584104],
    [40.99952629046706, 28.977086852460396],
    [40.81218286407427, 29.336889080006245],
    [40.9030665943093, 29.4179132457513],
    [41.065825281469394, 29.244191941230117],
    [41.05288150660772, 29.125402274502193],
    [41.167733090931044, 29.099996392022817],
    [41.23076416873252, 29.023092096563687],
    [41.113436436988906, 28.978460140856665],
    [41.199256217894124, 28.89331610227711],
    [41.259159966094586, 28.715474921912715],
    [41.182204608274056, 28.701055366991987],
    [41.08497739498677, 28.57745918195715]
]

city_polygon = Polygon(istanbul_coords)

# Grid ayarları
lat_offset = 0.01  # yaklaşık 1 km
lon_offset = 0.01

# Bounding box
min_lat = min(lat for lat, lon in istanbul_coords)
max_lat = max(lat for lat, lon in istanbul_coords)
min_lon = min(lon for lat, lon in istanbul_coords)
max_lon = max(lon for lat, lon in istanbul_coords)

# Grid ile aday istasyonlar
candidate_stations = []
lat = min_lat
while lat < max_lat:
    lon = min_lon
    while lon < max_lon:
        point = Point(lat + lat_offset/2, lon + lon_offset/2)
        if city_polygon.contains(point):
            candidate_stations.append([point.x, point.y])
        lon += lon_offset
    lat += lat_offset

print(f"Aday istasyon sayısı: {len(candidate_stations)}")

# --- Folium ile görselleştirme ---
avg_lat = (min_lat + max_lat)/2
avg_lon = (min_lon + max_lon)/2
m = folium.Map(location=[avg_lat, avg_lon], zoom_start=10, tiles="CartoDB positron")

# İstanbul poligonu
folium.Polygon(
    locations=istanbul_coords,
    color="blue",
    weight=3,
    fill=True,
    fill_opacity=0.2,
    popup="İstanbul Sınırı"
).add_to(m)

# Aday istasyon noktaları
for s in candidate_stations:
    folium.CircleMarker(
        location=[s[0], s[1]],
        radius=3,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.7
    ).add_to(m)

# Haritayı kaydet
m.save("istanbul_candidate_stations.html")
print("Harita kaydedildi: istanbul_candidate_stations.html")
```
